```python
# treatment of outliers

cols_to_treat = ['Weekly_Sales','Temperature', 'Unemployment']

for i in cols_to_treat:
  Q1 = df[i].quantile(0.25)
  Q3 = df[i].quantile(0.75)
  IQR = Q3 - Q1
  UL = Q3 + 1.5*IQR
  LL = Q1 - 1.5*IQR
  df = df[(df[i]>=LL) & (df[i]<=UL)]
```

```python
# check for outliers:

plt.figure(figsize=(10,10))
for i in range(0, len(df.columns)):
  plt.subplot(4,2,i+1)
  sns.boxplot(df[df.columns[i]],color='r')
  plt.title(df.columns[i])
  plt.tight_layout()

plt.show()
```

b. If the weekly sales show a seasonal trend, when and what could be the reason?

```python
df.head()
```

```python
df.tail()
```

```python
b = df.groupby('Date')['Weekly_Sales'].sum()
b
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(15,10))
plt.plot(b)
plt.title("Weekly Sales Trend")
plt.xlabel("date")
plt.ylabel("Weekly Sales")
plt.show()
```

# **Insights:**

- As we can clearly see that there is a seasonal trend present in the weekly Sales

- There is some same seasonal pattern during the Weeks of `Feb to May` , `May to December` and same pattern is falling in the Week of `January` month


- There might be the several resons for this trend: Customers may be regularly purchasing during `Feb to May` to fullfill their daily needs and

- there might be other reason that consumers are having busy schedule in `Jan` the reason behind there is a quick fall in the trend

2. Use predictive modeling techniques to forecast the sales for each store for the next 12
weeks.

```python
df.tail()
```

```python
# seggregation or filtering the data into each store data:

store_1 = df[df['Store']==1]
store_11 = df[df['Store']==11]
store_23 = df[df['Store']==23]
store_35 = df[df['Store']==35]
store_45 = df[df['Store']==45]
```

```python
store_1
```

```python
store_1 = store_1[['Date','Weekly_Sales']]
store_1
```

**Installing the Model library**

```python
from prophet import Prophet
```

**renaming the dataframe columns into 'ds' 'y' format since prophet model can only take these 2 columns**

```python
store_1.rename(columns={'Date':'ds','Weekly_Sales':'y'},inplace=True)
```

```python
store_1.head()
```

```python
# fitting the data into the model:

model_1 = Prophet(interval_width=0.95)
model_1.fit(store_1)
```

## forecasting the sales for next 12 weeks:

```python
store_1_future = model_1.make_future_dataframe(periods=84)
store_1_future
```

```python
store_1_forecast = model_1.predict(store_1_future)
store_1_forecast[['ds','yhat_lower', 'yhat', 'yhat_upper']].tail(84)
```

```python
# visualisation of predictions:

store_1_forecast_plot = model_1.plot(store_1_forecast)
plt.title("Store 1 Sales Forecast")
plt.show()
```
